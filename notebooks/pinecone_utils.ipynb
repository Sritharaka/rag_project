```python
# pinecone_utils.py
import os
import sys
sys.path.append(os.getcwd())  # Makes sure current folder is in path


from pinecone import Pinecone, ServerlessSpec

def init_pinecone():
    pc = Pinecone(api_key=PINECONE_API_KEY)

    # Extract region from PINECONE_ENVIRONMENT like 'gcp-starter' -> 'starter'
    region = PINECONE_ENVIRONMENT.split("-")[-1]
    cloud = PINECONE_ENVIRONMENT.split("-")[0]

    # Create index if it doesn't exist
    if PINECONE_INDEX_NAME not in pc.list_indexes().names():
        pc.create_index(
            name=PINECONE_INDEX_NAME,
            dimension=VECTOR_DIM,
            metric="cosine",
            spec=ServerlessSpec(cloud=cloud, region=region)
        )

    return pc.Index(PINECONE_INDEX_NAME)

def upsert_vectors(index, vectors):
    index.upsert(vectors=vectors)

```
